```python
from pydantic import BaseModel

class basic_input(BaseModel):
    kspacing: float = 0.04
    encut: float = 500
    
class NVTInputT(basic_input):
    scf_thr: float = 1e-6
    temp_begin: float = 300
    temp_end: float = 300

    parameters: str = ''
    
a= NVTInputT()
print(a.kspacing)
```
